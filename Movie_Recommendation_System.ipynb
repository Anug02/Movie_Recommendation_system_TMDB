## Method 1:Demographic Filtering
Here,the movie is shortlisted based on **IMDB Rating Formula** 

### IMDB Weighted Rating Formula

IMDB uses a weighted average to calculate its Top 250 rankings to ensure that movies with thousands of votes are prioritized over those with only a few.

The formula is defined as:

$$W = \frac{v}{v + m} R + \frac{m}{v + m} C$$

**Where:**
* **$W$**: Weighted Rating
* **$R$**: Average rating for the movie (from 1 to 10)
* **$v$**: Number of votes for the movie
* **$m$**: Minimum votes required to be listed in the Top 250
* **$C$**: The mean vote across the whole report

```python
C =df2['vote_average'].mean()
C
```

```python
m =df2['vote_count'].quantile(0.9)
m
```

```python
q_movies =df2.copy().loc[df2['vote_count']>=m]
q_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v =x['vote_count']
    R =x['vote_average']

    return (v/(v+m) *R) +(m/(m+v) *C)
```

```python
q_movies['score'] =q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies =q_movies.sort_values('score',ascending=False)
q_movies[['title','vote_count','vote_average','score']].head()
```

## Content Based Filtering

```python
df2['overview'].head(5)

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf =TfidfVectorizer(stop_words="english")
df2['overview'] =df2['overview'].fillna('')

tfidf_matrix =tfidf.fit_transform(df2['overview'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim =linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim[1]
```

```python
indices =pd.Series(df2.index, index=df2['title']).drop_duplicates()
indices
```

```python
type(indices)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    
    idx =indices[title]

    sim_scores =list(enumerate(cosine_sim[idx]))
    sim_scores =sorted(sim_scores, key=lambda x:x[1], reverse=True)

    sim_scores =sim_scores[1:11]
    movies_indices =[i[0] for i in sim_scores]

    return df2['title'].iloc[movies_indices]
    
```

```python
get_recommendations("The Avengers")
```

## Credits,Genres and Keywords Based Recommender

```python
from ast import literal_eval

features =['cast','crew','keywords','genres']
for feature in features:
    df2[feature] =df2[feature].apply(literal_eval)
```

```python
def get_director(x):
    for i in x:
        if i['job'] =='Director':
            return i['name']

    return ''
```

```python
def get_list(x):
    if isinstance(x, list):
        names =[i['name'] for i in x]
        if len(names)>3:
            names = names[:3]

        return names

    return []
```

```python
df2['director'] =df2['crew'].apply(get_director)

features =['cast','keywords','genres']
for feature in features:
    df2[feature] =df2[feature].apply(get_list)
```

```python
def clean_data(x):
    if isinstance(x,list):
        return [str.lower(i.replace(" ","")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ",""))

        else:
            return ''
```

```python
new_features =['cast','keywords','director','genres']
for feature in features:
    df2[feature] =df2[feature].apply(clean_data)
```

now,we are going to add all this features for creating a one entity.

```python
def one_entity(x):
    return str.join(x['keywords'])+''+str.join(x['cast'])+''+x['director']+str.join(x['genres'])

df2['one'] =df2.apply(one_entity, axis=1)
```

```python
df2["cast"] =df2["cast"].str.join(',')
df2["cast"]
```

```python
df2["keywords"] =df2["keywords"].str.join(',')
df2["genres"] =df2["genres"].str.join(',')
df2["one"] =df2["cast"] +" "+df2["keywords"] +" "+df2["director"]+" " +df2["genres"]
df2["one"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count =CountVectorizer(stop_words="english")
count_matrix =count.fit_transform(df2["one"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim =cosine_similarity(count_matrix, count_matrix)

df2 =df2.reset_index()
indices =pd.Series(df2.index, index=df2['title'])
```
